```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv("022bim024.csv")    #must provide path containing your name or roll number
# To display the top 5 rows 
df.head(8)
```

```python
df.tail(5)                        # To display the botton 5 rows
```

```python
df.dtypes
```

```python
df = df.drop(['fuel_type', 'body_style', 'wheel_base', 'engine_type', ], axis=1)
df.head(8)
```

```python
#Renaming the columns
df = df.rename(columns={"number_of_doors": "Doors", "drive_wheels": "wheels", "number_of_cylinders": "cylinders", "engine_location": "location", "price": "Price" })
df.head(8)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()  
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())
```

```python
# Plotting a basic histogram
plt.hist(x=df['make'], bins=30, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
 
# Display the plot
plt.show()
```

```python


df['make'].value_counts().nlargest(40).plot(kind='bar', figsize=(10, 5))

plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make")

plt.show()
```

```python
sns.boxplot(x=df['Price'])
plt.title("Boxplot of Price")
plt.show()
```

```python
sns.boxplot(x=df['horsepower'])
plt.show()
```

```python
sns.boxplot(x=df['width'])
plt.show()
```

```python
selected_columns = ['horsepower', 'engine_size', 'width', 'city_mpg', 'city_mpg', 'price']
df_selected = df[selected_columns]


plt.figure(figsize=(15, 10))

# Using Seaborn to create a heatmap
sns.heatmap(df_selected.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)

plt.title('Correlation Heatmap')
plt.show()
```

```python
#scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['horsepower'], df['horsepower'])
ax.set_xlabel('horsepower')
ax.set_ylabel('price')
plt.show()
```
